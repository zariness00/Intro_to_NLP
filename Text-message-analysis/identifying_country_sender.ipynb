The second objective is to use MultinomialNB to identify from which country a message has been sent to

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
import pandas as pd
```

```python
bow_vectorizer = CountVectorizer()
```

```python
df = pd.read_csv('processed_cleaned_sms.csv')
df.head()
```

```python
df["country"].value_counts()
```

```python
countries_over100 = df["country"].value_counts()[df["country"].value_counts() > 100].index

countries_over100
```

```python
df = df[df['country'].isin(countries_over100)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
X = df["Message_lemmatized"]
y = df["country"]
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 42)
```

```python
type(test_data)
```

```python
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(train_data)
X_test_bow = bow_vectorizer.transform(test_data)
```

```python
message_classifier = MultinomialNB()
```

```python
message_classifier.fit(X_train_bow, train_labels)
```

```python
predictions = message_classifier.predict(X_test_bow)
```

```python
from sklearn.metrics import classification_report
```

```python
print("Accuracy:", accuracy_score(test_labels, predictions))
print("\nClassification Report:\n", classification_report(test_labels, predictions))
```
